```python
import pandas as pd
import csv
```

```python
with open('/home/yhbedoya/Drive/Data Science and Engineering - PoliTo2/Thesis/AudiosetDownloader/data/unbalanced_train_segments.csv') as infile:
    reader = csv.reader(infile)
    next(reader)
    rows = [[row[0], row[1], row[2], row[3:]] for row in reader]
del rows[:2]   
```

```python
labelsDf = pd.DataFrame(rows[1:], columns=['#YTID', 'start_seconds', 'end_seconds', 'positive_labels'])
labelsDf
```

```python
def replacer(row):
    line = []
    for x in row:
        line.append(x.replace('"', ''))

    return line


labelsDf['positive_labels'] = labelsDf['positive_labels'].apply(replacer)
labelsDf
```

```python
labels = {
    'BassDrum': ["/m/0bm02"],
    'Rail transport':["/m/06d_3", "/m/07jdr", "/m/01g50p", "/t/dd00048", "/m/0195fx"],
    'Water': ["/m/0838f","/m/06mb1", "/m/0j6m2", "/m/0j2kx", "/m/05kq4", "/m/06wzb", "/m/07swgks"],
    'Dog':["/m/0bt9lr", "/m/05tny_", "/m/07r_k2n", "/m/07qf0zm", "/m/07rc7d9", "/m/0ghcn6", "/t/dd00136", "/m/07srf8z"],
    'Siren':["/m/03kmc9", "/m/04qvtq", "/m/012n7d", "/m/012ndj", "/m/0dgbq"],
    'Tools': ["/m/07k1x", "/m/03l9g", "/m/03p19w", "/m/01b82r", "/m/02p01q", "/m/023vsd", "/m/0_ksk"],
    'Wind':["/m/03m9d0z","/m/07q8f3b", "/m/09t49", "/t/dd00092"],
    "Bird vocalization, bird call, bird song":["/m/020bb7","/m/07pggtn", "/m/07sx8x_"],
    'Fowl':["/m/025rv6n","/m/09b5t", "/m/01rd7k", "/m/09ddx", "/m/0dbvp"]

}
```

```python
def checker(row):
    match = []
    for slot in reversed(row):
        for k,v in labels.items():
            if slot in v:
                return [True, k]
    return [False, 'Nan']
```

```python
pack = labelsDf['positive_labels'].apply(checker)
pack = pd.DataFrame(pack.tolist(), columns=['mask', 'folder'])
pack
```

```python
labelsDf['mask'] = pack['mask'].tolist()
labelsDf['folder'] = pack['folder'].tolist()
usefulDf = labelsDf[labelsDf['mask']]
usefulDf
```

```python
usefulDf['folder'].value_counts()
```

```python
usefulDf['fileName'] = (usefulDf['#YTID'] + '_' +usefulDf['start_seconds'].astype(str) + '.wav').replace(' ', '')
usefulDf['fileName'] = usefulDf['fileName'].apply(lambda x: x.replace(' ', ''))
usefulDf
```

```python
zipText = {}

for k, v in labels.items():
    slice = usefulDf[usefulDf['folder']==k]
    text = 'zip ' + k + '.zip'
    for filename in slice['fileName'].tolist():
        text = text + "," + filename

    zipText[k]= text
```
